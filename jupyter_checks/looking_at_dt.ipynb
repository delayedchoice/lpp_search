```python

def make_LightKurveObject(time, flux, flux_err):
    """ Convert this object to a lightkurve.lightcurve.TessLightCurve object to use for Deep Transit.

    :return: Data in a lightkurve object
    :rtype: `class` lightkurve.lightcurve.TessLightCurve
    """
    lc = lk.TessLightCurve()
    lc.time = time
    lc.flux = flux
    lc.flux_err = flux_err
    return lc

df = pd.read_csv(glob.glob('../toi_data/*'+str(ticid)+'*/*8_TGLC_APER_total.csv')[0])

lc1 = make_LightKurveObject(df['TIME'], df['RAW_FLUX'], df['RAW_FLUX_ERR'])
```

```python
bboxes = dt.DeepTransit(lc).transit_detection('../model_TESS.pth', confidence_threshold=0.02)
bboxes1 = dt.DeepTransit(lc1).transit_detection('../model_TESS.pth', confidence_threshold=0.02)
```
