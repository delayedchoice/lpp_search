```python
from pathlib import Path

def find_error_in_files(directory_path, search_word="Error", incl_str = ''):
    """
    Scans through all .txt files in the specified directory and its subdirectories
    for a specific word and prints the filenames that contain it.

    Args:
        directory_path (str): The path to the directory to scan.
        search_word (str): The word to search for (case-sensitive).
    """
    # Create a Path object for the directory
    path = Path(directory_path)
    
    keep_list = []
    # Use rglob to find all files ending with .txt recursively
    for file_path in path.rglob('test_search_w_plots'+incl_str+'*.txt'):
        if file_path.is_file():
            try:
                # Open the file and read its content
                content = file_path.read_text(encoding='utf-8')
                
                # Check if the search word is in the content
                if (search_word in content):
                    print(f"Found '{search_word}' in file: {file_path}")
                    keep_list.append(str(file_path))

                elif 'slurmstepd: error:' in content:
#                     print(f"Found cancelled '{search_word}' in file: {file_path}")
                    continue


            except UnicodeDecodeError:
                print(f"Could not read file {file_path} due to encoding error, skipping.")
            except Exception as e:
                print(f"An unexpected error occurred while reading {file_path}: {e}")
    return keep_list

# --- Example Usage ---
# Replace '.' with the path to your specific directory, e.g., 'C:\\Users\\YourUser\\Logs'
# or use '.' to scan the current directory where the script is located.
scan_directory = '../carc/SLURM_output/'#test_search_w_plots31*' 
print(f"Starting scan in: {Path(scan_directory).resolve()}")
double_check = find_error_in_files(scan_directory)
```
